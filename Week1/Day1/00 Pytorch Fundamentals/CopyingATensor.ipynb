### Copying 
a crutial in Deep learning sometime when you want to preserve some tensor or to use the same tensor for two different we create a 
 copy of that tensor 

We can create the copy of tensor by simple assignment operator but that doesnt help much.
as when using the assignment operator the variable points to the same memoery address where it was previously created. 

This cabe displayed as follows 

```python
import torch
a =  torch.rand(2,2)
b = a 
print("memoery address of a an be given respectively ")
print(id(a))
id(b)


print("This means that if any change will be done to the a tensor that would be reflected on the b tensor as well")

a +=1

print(a)
print(b)
```

## To avoid this we can use the clone() to avoid this and create a separate memory of the the tensor at a differnt location in the memory 

```python
c = a.clone()
print("Now the memory address of c would be different from a and b since we used the clone method")
print(id(c))
print(id(a))
```

```python
## Any change to the a will not affect the c tensor but will affect the b tensor since it is jus the reference of a

a+=1
print(a)
print(c)
```
