```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

# For Categorical Data

# 1. Count Plot



```python
sns.countplot(df['Sex'])
```

```python
sns.countplot(df['Pclass'])
```

```python
df.Pclass.dtype ## the above plot appeared like that because i didnt use the x parameter of the countplot function

### When you pass a Series without specifying x=, seaborn:

# Uses the index as categories

# Each index value appears once

# Result → hundreds of thin bars (exactly what you’re seeing)
```

```python
sns.countplot(x='Pclass', data=df)
```

# B.  Pie Chart


```python
df['Survived'].value_counts().plot(kind = "pie") 
```

```python
# we can us eht autopct parameter to show the percentage of each class 

df['Survived'].value_counts().plot(kind="pie",autopct = '%0.2f')
```

# Numerical Data 
 what kind of plot can be made if the data is numerical 

```python
import matplotlib.pyplot as plt

## ploting the histogram using matplotlib


# histogrma :- enable us to see the distribution of numerical data 
# we can modify the bins paraeter to change the number of bars in the histogram 
plt.hist(df["Age"], bins=20)
```

```python
# Displot = sns.histplot() + kde line 
# it combines both histogram and kde plot used  for understanding the distribution of numerical data

sns.displot(df['Age'], kde=True, bins=20)
```

```python
# Box plot

sns.boxplot(y = df['Age'])
```
