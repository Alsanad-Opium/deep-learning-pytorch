```python
## this skimpy library is a great alternative to sweetviz for EDA. It provides a concise summary of the dataset, including data types, missing values, and basic statistics. It also generates a report that can be easily shared with others.

## This code reads the Titanic dataset and uses the skimpy library to generate a summary report of the dataset. The skim function provides insights into the data types, missing values, and basic statistics of the dataset, which can help in understanding the data better before performing any analysis or modeling.
```
