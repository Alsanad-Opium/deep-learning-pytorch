```python
## to load all the data we can rrun this script 
 
url  = "https://api.imdbapi.dev/titles"
all_data = []
MAX_pages = 3
page = 0
while page < MAX_pages:
    
    res = requests.get(url)
    data = res.json()
    
    all_data.extend(data['titles'])
    
    next_page = data.get('nextPageToken')
    page +=1
    print(page)
    if next_page:
        url = f"https://api.imdbapi.dev/titles?nextPageToken={next_page}"
    else:
        break

df = pd.DataFrame(all_data)

df.head()
df.shape
```

```python
df.to_csv("fetched_movie_data.csv")
```
