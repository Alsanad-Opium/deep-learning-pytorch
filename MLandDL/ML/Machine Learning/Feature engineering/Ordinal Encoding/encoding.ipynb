```python
import pandas as pd
df = pd.read_csv('customer.csv')
df.sample(5)

```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

df['review'] = oe.fit_transform(df[['review']])

df['review'].head()
```

```python
df['review'].unique()
```

```python
mapping = dict(zip(oe.categories_[0], range(len(oe.categories_[0]))))

# oe.categories_[0] → gives category names
# Example: ['average', 'bad', 'good']

# range(len(...)) → creates numbers
# Example: [0, 1, 2]

# zip() → joins them pair-by-pair

# ('average', 0)
# ('bad', 1)
# ('good', 2)

# dict() → converts pairs into dictionary

# {'average': 0, 'bad': 1, 'good': 2}
```
