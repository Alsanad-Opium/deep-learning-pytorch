```python
## as you can see in the output we have Four new columns for each unique value in the fuel column and we have 1 in the column that corresponds to the value in the fuel column and 0 in the other columns.

## Now to avoid multicollinearity we will drop one of the columns and we will drop the column that corresponds to the most frequent value in the fuel column which is petrol.

## this can be done as follows

pd.get_dummies(df, columns=['fuel', 'owner'], dtype=int, drop_first=True)
```

```python
## now we will do the same using the fuel column using the sklearn library # first we have split the data in to x_train x_test and Y_train y_test 

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop('selling_price',axis =1),df.selling_price, test_size = 0.2,random_state =42)

```

```python
## Now we will create the object of ONEHOTENCCODER 

ohe = OneHotEncoder(drop = 'first')

x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
x_test_new = ohe.transform(x_test[['fuel','owner']])

print(x_train_new.shape)
print(x_test_new.shape)
print(x_train.head())
```

```python
import numpy as np 
```

```python
df_new= pd.DataFrame(np.hstack((x_train[['brand','km_driven']].values, x_train_new.toarray())),index=x_train.index, columns=['brand','km_driven']+list(ohe.get_feature_names_out()))


## this above is how we can do one hot encoding using the sklearn library and we have to convert the sparse matrix to a dense matrix using the toarray() method and then we have to concatenate it with the original dataframe using the hstack() method from numpy and then we have to create a new dataframe from the concatenated array and we have to set the index and columns of the new dataframe.

## in simple terms the above code is doing the same thing as the get_dummies function from pandas but we have to do it manually using the OneHotEncoder class from sklearn and we have to convert the sparse matrix to a dense matrix and then we have to concatenate it with the original dataframe and then we have to create a new dataframe from the concatenated array and we have to set the index and columns of the new dataframe.
```
